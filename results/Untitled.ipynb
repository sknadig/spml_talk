```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
phn_data = np.load("FADG0_SI1909.ep.16_phn.npy")
char_data = np.load("FADG0_SI1909.ep.16_char.npy")
```

```python
phn_data.shape, char_data.shape
```

```python
# plt.figure(figsize=(20,10))
j=33
k=48
labels = ["" for _ in range(k-j)]
labels[3] = "R"
labels[5] = "O"
labels[7] = "L"
labels[9] = "L"
labels[11] = "S"

fig, ax1 = plt.subplots(figsize=(30,15))

l_w=8
ax1.plot(char_data[20, j:k], color="red", label="R", lw=l_w)
ax1.plot(char_data[21, j:k], color="green", label="O", lw=l_w)
ax1.plot(char_data[22, j:k], color="blue", label="L", lw=l_w)
ax1.plot(char_data[23, j:k], color="blue", label="L", lw=l_w)
ax1.plot(char_data[24, j:k], color="magenta", label="S", lw=l_w)
# ax[0].plot(char_data[9, :k], color="black")

ax1.plot(phn_data[13, j:k], '--', color="red", label="R", lw=l_w)
ax1.plot(phn_data[14, j:k], '-.', color="green", label="OW", lw=l_w)
ax1.plot(phn_data[15, j:k], ':', color="blue", label="L", lw=l_w)
ax1.plot(phn_data[16, j:k], '--', color="magenta", label="Z", lw=l_w)
ax1.set_ylabel("Attention weight", fontsize=40)
ax1.set_xlabel("Input sequence", fontsize=40)
ax1.set_xticks(np.arange(k-j))
ax1.set_xticklabels(labels)
ax1.tick_params(labelsize=70)
ax1.legend(fontsize=40)


labels2 = ["" for _ in range(k-j)]

labels2[3] = "R"
labels2[5] = "OW"
labels2[10] = "L"
labels2[13] = "Z"

ax2 = ax1.twiny()
ax2.set_xlim(0,17)
ax2.set_xticks(np.arange(k-j))
ax2.set_xticklabels(labels2)
ax2.tick_params(labelsize=70)

# ax2.axvline(x=3, ls='--', color="black", alpha=0.3)
# ax2.grid()

plt.savefig("att_ind.eps", format="eps", dpi=300)
```
